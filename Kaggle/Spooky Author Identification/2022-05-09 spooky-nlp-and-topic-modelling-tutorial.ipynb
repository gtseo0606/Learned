# 2. Natural Language Processing

![](https://s-media-cache-ak0.pinimg.com/originals/d8/ca/04/d8ca049fd1ad2d92818763e55c57f49a.jpg)

거의 모든 자연어 처리(컴퓨터와 인간 언어 간의 상호 작용을 탐구하는 분야) 작업(주제 모델링, 단어 클러스터링 또는 문서 텍스트 분류 등)에서 일반적으로 항상 이러한 작업을 거쳐야 합니다. 입력된 원시 텍스트를 모델과 기계가 읽을 수 있는 형식으로 변환하기 위한 몇 가지 전처리 단계. Random Forest 모델에 단어의 단락을 제공하고 해당 단락이 어느 저자에게서 왔는지 즉시 예측할 것으로 기대할 수는 없습니다. 이면에서 텍스트 사전 처리는 다음과 같은 몇 가지 간단한 단계로 요약될 수 있습니다.

1. **토큰화** - 텍스트를 개별 구성 단어로 분리합니다.
2. **중단어** - 너무 자주 발생하는 단어는 발생 빈도가 관련 텍스트 감지에 도움이 되지 않으므로 버리십시오. (제쳐두고 매우 드물게 발생하는 단어를 버리는 것도 고려하십시오).
3. **형태소 분석** - 단어의 변형을 여전히 동일한 의미를 전달하는 단일 상위 단어로 결합합니다.
4. **벡터화** - 텍스트를 벡터 형식으로 변환합니다. 가장 간단한 방법 중 하나는 행렬(코퍼스의 각 문서 또는 텍스트에 대해)을 만드는 유명한 단어 모음 접근 방식입니다. 가장 단순한 형태로, 이 행렬은 단어 빈도(단어 수)를 저장하고 원시 텍스트의 벡터화라고 합니다.

**Natural Language Toolkit(NLTK)**: 자연어 처리 작업을 더욱 편리하게 하기 위해 NLP에서 가장 편리한 툴킷 중 하나인 Natural Language Toolkit(더 일반적으로 [NLTK]라고도 함)을 소개하겠습니다. ](http://www.nltk.org/) 모듈. 툴킷을 가져오려면 다음과 같이 더 쉽습니다.

## 2c. 형태소 분석 및 표제어 추출

불용어 제거 후, 제가 소개하고자 하는 NLP의 다음 단계는 Stemming 과정입니다. 이 단계의 작업은 유사한 단어의 다양한 변형을 단일 용어로 줄이기 위해 시도합니다(다른 분기는 모두 단일 단어 *줄기*로 축소됨). 따라서 "running", "runs" 및 "run"이 있는 경우 이 세 가지 고유한 단어가 "run"이라는 단어로 축소되기를 정말로 원할 것입니다. (그러나 물론 과거, 현재 또는 미래 시제의 세분성을 잃습니다).

Porter 형태소 분석 알고리즘, lancaster 형태소 분석기 및 Snowball 형태소 분석기와 같은 변형을 포함하는 다양한 형태소 분석기를 제공하는 NLTK로 다시 전환할 수 있습니다. 다음 예에서는 다음과 같이 포터 형태소 분석기 인스턴스를 생성합니다.

```python
stemmer = nltk.stem.PorterStemmer()
```

이제 형태소 분석기를 사용하여 이러한 테스트 단어("running", "runs", "run")를 단일 형태소 단어로 줄일 수 있는지 확인할 수 있습니다. 편리하게 다음과 같이 즉시 형태소 분석기를 테스트할 수 있습니다.

```python
print("The stemmed form of running is: {}".format(stemmer.stem("running")))
print("The stemmed form of runs is: {}".format(stemmer.stem("runs")))
print("The stemmed form of run is: {}".format(stemmer.stem("run")))
```

우리가 볼 수 있듯이, 형태소 분석기는 위의 주어진 단어를 기본 형식으로 성공적으로 축소했으며 이는 학습 및 분류 작업에 올 때 단어 데이터 세트의 크기를 줄이는 데 가장 도움이 될 것입니다.

그러나 형태소 분석에는 한 가지 결함이 있으며 그 과정에는 [단어 끝을 자르는 조잡한 휴리스틱](https://nlp.stanford.edu/IR-book/html/htmledition/stemming- and-lemmatization-1.html) 특정 단어를 사람이 인식할 수 있는 기본 형식으로 줄이기를 희망합니다. 따라서 이 프로세스는 이 예에서 설명하는 것처럼 단어를 축소할 때 어휘나 단어 형식을 고려하지 않습니다.

불행히도 벡터라이저에는 내장형 표제어가 없으므로 몇 가지 옵션이 남아 있습니다. 벡터화를 위해 데이터를 공급하기 전에 매번 별도로 구현하거나 이 기능을 포함하도록 sklearn 구현을 확장합니다. 운 좋게도 다음과 같이 "build_analyzer" 메서드를 덮어써서 CountVectorizer 클래스를 확장할 수 있는 후자의 옵션이 있습니다.

### 표제어로 CountVectorizer 클래스 확장하기

```python
'''class 제작클래스(A함수):
    def 제작함수(b):
        제작변수 = super(제작클래스, b).제작함수
        return lambda doc : (B함수 for w in 제작변수(doc))'''
```

```python
class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc : (WordNetLemmatizer().lemmatize(w) for w in analyzer(doc))
```

여기서는 객체 지향 프로그래밍(OOP)의 미묘한 개념을 활용했습니다. 기본적으로 원본 Sklearn의 CountVectorizer 클래스를 상속 및 하위 클래스화했으며 원시 텍스트 매트릭스의 각 목록에 대한 표제어를 구현하여 build_analyzer 메서드를 덮어썼습니다.

```python
# Storing the entire training text in a list
text = list(train.text.values)

# Calling our overwritten Count vectorizer
tf_vectorizer = LemmaCountVectorizer(max_df=0.95, min_df=2, stop_words='english', decode_error='ignore')
tf = tf_vectorizer.fit_transform(text)
```

## 3b. 잠재 디리클레 할당

마지막으로 주제 모델링과 몇 가지 비지도 학습 알고리즘의 구현이라는 주제에 도달합니다. 첫 번째로 다룰 방법은 [Latent Dirichlet Allocation](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.LatentDirichletAllocation.html)입니다. 이제 이 LDA 알고리즘의 두 가지 다른 구현이 있지만 이 노트북에서는 Sklearn의 구현을 사용할 것입니다. 또 다른 매우 잘 알려진 LDA 구현은 Radim Rehurek의 [gensim](https://radimrehurek.com/gensim/)이므로 확인하십시오.

**코퍼스 - 문서 - 워드 : 토픽 생성**

LDA에서 모델링 프로세스는 텍스트 말뭉치, 문서 모음, D 및 문서의 단어 W라는 세 가지를 중심으로 진행됩니다. 따라서 알고리즘은 다음 방법을 통해 이 말뭉치에서 K개의 주제를 찾으려고 시도합니다(다이어그램으로 표시됨).

![Three_Level 베이지안 모델](http://scikit-learn.org/stable/_images/lda_model_graph.png)

$\beta_{k}$가 제공하는 Dirichlet 사전 분포를 통해 $\kappa$ 각 주제를 모델링합니다.

![](http://scikit-learn.org/stable/_images/math/2c1ff5b3d6f342d7dad0395210c8a13947de451c.png)

$\alpha$로 매개변수화된 다른 Dirichlet 분포로 각 문서 d를 모델링합니다.

![](http://scikit-learn.org/stable/_images/math/530c80986933767c9d182af83075c13d72cdef97.png)

이어서 문서 d에 대해 다항 분포를 통해 주제를 생성한 다음 역추적하고 다른 다항 분포를 통해 해당 주제와 관련된 해당 단어를 생성하는 데 사용합니다.

![](http://scikit-learn.org/stable/_images/math/0bb078c0fe621366a147231b5c8240efacdb895b.png)
![](http://scikit-learn.org/stable/_images/math/f7960e06327c64a1e9da6769e68aa4342b7de73d.png)


*(이미지 출처: http://scikit-learn.org/stable/modules/decomposition.html#latentdirichletallocation)*

LDA 알고리즘은 먼저 혼합 주제 모델을 통해 문서를 모델링합니다. 그런 다음 이러한 주제에서 이러한 주제의 확률 분포를 기반으로 단어에 가중치가 할당됩니다. LDA 사용자가 특정 단어가 주제에 포함될 가능성을 말할 수 있도록 하는 것은 단어에 대한 이 확률적 할당입니다. 결과적으로 특정 주제에 할당된 단어 모음에서 우리는 해당 주제가 어휘적 관점에서 실제로 무엇을 나타낼 수 있는지에 대한 통찰력을 얻을 수 있습니다.

표준 LDA 모델에는 모델을 호출하기 전에 염두에 두고 프로그래밍 방식으로 조정하는 것을 고려해야 하는 몇 가지 주요 매개변수가 있습니다.
1. n_components: 모델에 지정하는 주제의 수
2. $\alpha$ 매개변수: 이전 문서 주제에 연결할 수 있는 dirichlet 매개변수입니다.
3. $\beta$ 매개변수: 이는 사전 주제어에 연결된 디리클레 매개변수입니다.

알고리즘을 호출하려면 Sklearn의 *LatentDirichletAllocation* 함수를 통해 LDA 인스턴스를 생성하기만 하면 됩니다. 다양한 매개변수는 일종의 검증 체계를 통해 이상적으로 얻어졌을 것입니다. 이 경우 n_components(또는 주제 번호)의 최적 값은 Kmeans 클러스터의 수와 LSA 차원의 수를 반복하고 가장 좋은 KMeans + Latent Semantic Analysis Scheme(이 백서에 표시된 대로)을 수행하여 찾았습니다. 실루엣 평균 점수입니다.

```python
lda = LatentDirichletAllocation(n_components=11, max_iter=5,
                                learning_method = 'online',
                                learning_offset = 50.,
                                random_state = 0)
```

```python
lda.fit(tf) # tf=희소행렬
```
